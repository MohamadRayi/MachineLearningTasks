```python
# Mengimpor pustaka PyTorch
import torch

# Mengimpor kelas optimizer AdamW dan kelas-kelas terkait dari pustaka Transformers
from transformers import AdamW, AutoTokenizer, AutoModelForSequenceClassification

# Menentukan model yang akan digunakan
# Model yang dipilih adalah 'bert-base-uncased', yaitu model BERT dengan teks huruf kecil semua
checkpoint = "bert-base-uncased"

# Menginisialisasi tokenizer berdasarkan model yang telah ditentukan sebelumnya
# Tokenizer ini digunakan untuk mengubah teks menjadi token yang dapat diproses oleh model BERT
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

# Menginisialisasi model untuk klasifikasi urutan teks berdasarkan model yang telah ditentukan sebelumnya
# Model ini digunakan untuk tugas klasifikasi dengan jumlah label yang sesuai dengan tugasnya
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)

# Menyiapkan urutan teks yang akan diproses
sequences = [
    "I've been waiting for a HuggingFace course my whole life.",
    "This course is amazing!",
]

# Membuat batch dari urutan teks dengan padding dan truncation
# Batch ini akan digunakan sebagai input untuk model, dengan padding untuk memastikan panjang yang sama dan truncation untuk memotong teks yang terlalu panjang
batch = tokenizer(sequences, padding=True, truncation=True, return_tensors="pt")

# Menambahkan label-label yang sesuai ke dalam batch
# Label ini digunakan untuk tujuan supervisi selama pelatihan, di mana 1 mungkin menunjukkan kelas positif dalam konteks ini
batch["labels"] = torch.tensor([1, 1])

# Menginisialisasi optimizer AdamW dengan parameter-parameter model
# AdamW adalah varian dari Adam optimizer yang menggabungkan regularisasi L2 weight decay untuk mengurangi overfitting
optimizer = AdamW(model.parameters())

# Menghitung loss dan melakukan backpropagation
# Forward pass: model menerima input batch dan menghasilkan output, loss dihitung berdasarkan perbedaan antara prediksi model dan label yang benar
loss = model(**batch).loss

# Backward pass: menghitung gradien dari loss sehubungan dengan parameter model
loss.backward()

# Melakukan satu langkah optimisasi (update parameter)
# Optimizer mengupdate parameter model untuk meminimalkan loss berdasarkan gradien yang dihitung
optimizer.step()
```

```python
# Mengimpor fungsi yang diperlukan dari pustaka datasets
from datasets import load_dataset

# Memuat dataset MRPC (Microsoft Research Paraphrase Corpus) dari dataset GLUE
raw_datasets = load_dataset("glue", "mrpc")

# Menampilkan dataset yang dimuat
raw_datasets
```

```python
# Mengimpor kelas AutoTokenizer dari pustaka Transformers
from transformers import AutoTokenizer

# Menentukan titik kontrol yang akan digunakan untuk tokenizer
checkpoint = "bert-base-uncased"

# Membuat objek tokenizer menggunakan model BERT "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

# Melakukan tokenisasi pada kalimat-kalimat dari dataset pelatihan (train)
tokenized_sentences_1 = tokenizer(raw_datasets["train"]["sentence1"])
tokenized_sentences_2 = tokenizer(raw_datasets["train"]["sentence2"])
```

```python
# Tokenizer digunakan untuk memproses teks menjadi token-token yang dapat dipahami oleh model
inputs = tokenizer("This is the first sentence.", "This is the second one.")

# 'inputs' berisi hasil tokenisasi dari dua kalimat yang telah dimasukkan
```

```python
# Mengonversi token-token dalam bentuk ID menjadi token-token teks menggunakan tokenizer yang telah didefinisikan sebelumnya
tokenizer.convert_ids_to_tokens(inputs["input_ids"])
```

```python
# Melakukan tokenisasi pada dataset menggunakan tokenizer yang telah didefinisikan sebelumnya
tokenized_dataset = tokenizer(
    raw_datasets["train"]["sentence1"],  # Kalimat pertama dalam dataset pelatihan
    raw_datasets["train"]["sentence2"],  # Kalimat kedua dalam dataset pelatihan
    padding=True,   # Melakukan padding pada token agar memiliki panjang yang seragam
    truncation=True,  # Melakukan pemotongan token jika melebihi panjang maksimum
)
```

```python
# Mendefinisikan fungsi untuk tokenisasi data
def tokenize_function(example):
    # Menggunakan tokenizer untuk tokenisasi sentence1 dan sentence2 dari contoh
    return tokenizer(example["sentence1"], example["sentence2"], truncation=True)
```

```python

# Melakukan tokenisasi pada dataset menggunakan fungsi tokenize_function
# dengan pemrosesan berbasis batch (batched=True)
tokenized_datasets = raw_datasets.map(tokenize_function, batched=True)

# Menyimpan dataset yang telah ditokenisasi
tokenized_datasets
```

```python
# Mengimpor kelas DataCollatorWithPadding dari pustaka Transformers
from transformers import DataCollatorWithPadding

# Membuat objek DataCollatorWithPadding dengan tokenizer yang telah ditentukan sebelumnya
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
# Mengambil 8 contoh data dari dataset latihan
samples = tokenized_datasets["train"][:8]

# Menghapus kunci-kunci tertentu dari setiap contoh, seperti "idx", "sentence1", dan "sentence2"
samples = {k: v for k, v in samples.items() if k not in ["idx", "sentence1", "sentence2"]}

# Menghitung panjang dari setiap contoh data input berdasarkan token input_ids
[len(x) for x in samples["input_ids"]]
```

```python
# Memproses batch data menggunakan fungsi data_collator
batch = data_collator(samples)

# Membuat dictionary yang berisi informasi tentang bentuk (shape) dari setiap item dalam batch
{k: v.shape for k, v in batch.items()}
```
