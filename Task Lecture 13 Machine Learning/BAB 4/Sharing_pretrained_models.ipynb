```python
# Mengimpor kelas-kelas yang dibutuhkan dari pustaka Transformers
from transformers import AutoModelForMaskedLM, AutoTokenizer

# Menetapkan checkpoint dari model pra-terlatih yang akan digunakan
checkpoint = "camembert-base"

# Memuat model pra-terlatih untuk Masked Language Modeling dari checkpoint yang ditetapkan
model = AutoModelForMaskedLM.from_pretrained(checkpoint)

# Memuat tokenizer yang sesuai dengan model pra-terlatih dari checkpoint yang sama
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

# Menyimpan model yang telah dimuat ke folder yang dipilih
model.save_pretrained("./model")

# Menyimpan tokenizer yang telah dimuat ke folder yang dipilih
tokenizer.save_pretrained("./tokenizer")
```
