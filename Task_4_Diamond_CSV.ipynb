Mengimpor library yang diperlukan

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
```

Membaca data dari file CSV

```python
data = pd.read_csv(file_path)
```

Menampilkan Statistik fitur

```python
print("Statistik Fitur:")
stats = data.describe().transpose()
stats['mean'] = stats['mean'].map('{:.2f}'.format)
stats['std'] = stats['std'].map('{:.2f}'.format)
stats['min'] = stats['min'].map('{:.2f}'.format)
stats['25%'] = stats['25%'].map('{:.2f}'.format)
stats['50%'] = stats['50%'].map('{:.2f}'.format)
stats['75%'] = stats['75%'].map('{:.2f}'.format)
stats['max'] = stats['max'].map('{:.2f}'.format)
print(stats)
```

Pengaturan style dan warna

```python
sns.set_style("whitegrid")
sns.set_palette("pastel")
```

Pengaturan Ukuran Font

```python
sns.set_context("notebook", font_scale=1.2)
```

BoxPlot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, linewidth=2.5, notch=True)  # Mengatur ketebalan garis dan menampilkan notch
plt.title("Boxplot Data", fontsize=18, fontweight='bold', color='blue')  # Judul dengan ukuran font besar, tebal, dan warna biru
plt.xlabel("Features", fontsize=14, color='green')  # Label sumbu x dengan ukuran font dan warna hijau
plt.ylabel("Values", fontsize=14, color='green')  # Label sumbu y dengan ukuran font dan warna hijau
plt.xticks(rotation=45, fontsize=12, color='purple')  # Rotasi label sumbu x, ukuran font, dan warna label sumbu x
plt.yticks(fontsize=12, color='purple')  # Ukuran font dan warna label sumbu y
plt.grid(True, linestyle='--', linewidth=0.8, alpha=0.7, color='gray')  # Menampilkan grid dengan gaya putus-putus, lebar garis 0.8, dan transparansi 0.7
plt.show()
```

Distribusi Plot

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, kde=True, color='orange')  # Warna plot orange
plt.title("Distribusi Plot", fontsize=16, style='italic', color='red')  # Judul dengan ukuran font 16, gaya italic, dan warna merah
plt.xlabel("Values", fontsize=14, fontstyle='oblique', color='blue')  # Label sumbu x dengan ukuran font 14, gaya oblique, dan warna biru
plt.ylabel("Frequency", fontsize=14, fontstyle='oblique', color='blue')  # Label sumbu y dengan ukuran font 14, gaya oblique, dan warna biru
plt.grid(True, linestyle='-', linewidth=0.5, alpha=0.5, color='gray')  # Menampilkan grid dengan gaya garis solid, lebar garis 0.5, dan transparansi 0.5
plt.show()
```

Violin Plot

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, inner='quartile', palette='Set2')  # Menampilkan boxplot di dalam violin plot, menggunakan palet warna Set2
plt.title("Violin Plot", fontsize=20, fontstyle='normal', color='green')  # Judul dengan ukuran font 20, gaya normal, dan warna hijau
plt.xlabel("Features", fontsize=14, style='italic', color='purple')  # Label sumbu x dengan ukuran font 14, gaya italic, dan warna ungu
plt.ylabel("Values", fontsize=14, style='italic', color='purple')  # Label sumbu y dengan ukuran font 14, gaya italic, dan warna ungu
plt.grid(True, linestyle=':', linewidth=0.6, alpha=0.8, color='gray')  # Menampilkan grid dengan gaya garis titik-titik, lebar garis 0.6, dan transparansi 0.8
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="carat", y="price", data=data, marker='D', color='red')  # Mengatur marker diamond (D) dan warna merah
plt.title("Scatter Plot", fontsize=18, fontweight='bold', color='blue')  # Judul dengan ukuran font besar, tebal, dan warna biru
plt.xlabel("Carat", fontsize=14, style='italic', color='green')  # Label sumbu x dengan ukuran font 14, gaya italic, dan warna hijau
plt.ylabel("Price", fontsize=14, style='italic', color='green')  # Label sumbu y dengan ukuran font 14, gaya italic, dan warna hijau
plt.grid(True, linestyle='-', linewidth=0.5, alpha=0.5, color='gray')  # Menampilkan grid dengan gaya garis solid, lebar garis 0.5, dan transparansi 0.5
plt.show()
```

Pair Plot

```python
plt.figure(figsize=(12, 8))
sns.pairplot(data, vars=['carat', 'depth', 'price', 'x', 'y', 'z'], kind='scatter', diag_kind='kde')
plt.suptitle('Pairplot of Diamond Data', fontsize=20, fontweight='bold', color='blue')
plt.show()
```
