Import Library

```python
import pandas as pd  # Mengimpor Pandas untuk manipulasi data
import seaborn as sns  # Mengimpor Seaborn untuk visualisasi data
import matplotlib.pyplot as plt  # Mengimpor Matplotlib untuk fungsionalitas plotting tambahan
```

Load Dataset

```python
file_path = '/content/drive/MyDrive/Dataset/StudentsPerformance.csv'  # Path file dataset di Google Drive
data = pd.read_csv(file_path)  # Membaca dataset menjadi DataFrame Pandas
```

Set Style

```python
sns.set_context("notebook", font_scale=1.2)  # Mengatur konteks untuk plot Seaborn
sns.set_style("whitegrid")  # Mengatur gaya untuk plot Seaborn
```

Scatter plot

```python
plt.figure(figsize=(10, 6))  # Mengatur ukuran gambar
sns.scatterplot(data=data, x='math score', y='reading score', hue='gender', palette='Set2')  # Membuat scatter plot
plt.title('Scatter Plot Skor Matematika vs Skor Membaca', fontsize=16)  # Judul plot
plt.xlabel('Skor Matematika', fontsize=14)  # Label sumbu x
plt.ylabel('Skor Membaca', fontsize=14)  # Label sumbu y
plt.legend(fontsize=12)  # Legenda
plt.show()  # Menampilkan plot
```

Box Plot

```python
plt.figure(figsize=(10, 6))  # Mengatur ukuran gambar
sns.boxplot(data=data, x='parental level of education', y='math score', hue='parental level of education', palette='Pastel1', legend=False)  # Membuat box plot
plt.title('Box Plot Skor Matematika berdasarkan Tingkat Pendidikan Orang Tua', fontsize=16)  # Judul plot
plt.xticks(rotation=45, fontsize=12)  # Label sumbu x
plt.xlabel('Tingkat Pendidikan Orang Tua', fontsize=14)  # Label sumbu x
plt.ylabel('Skor Matematika', fontsize=14)  # Label sumbu y
plt.show()  # Menampilkan plot
```

Violin Plot

```python
plt.figure(figsize=(10, 6))  # Mengatur ukuran gambar
sns.violinplot(data=data, x='test preparation course', y='writing score', hue='gender', split=True, palette='Set3')
# Membuat violin plot
plt.title('Violin Plot Skor Menulis berdasarkan Kursus Persiapan Tes dan Jenis Kelamin', fontsize=16)  # Judul plot
plt.xlabel('Kursus Persiapan Tes', fontsize=14)  # Label sumbu x
plt.ylabel('Skor Menulis', fontsize=14)  # Label sumbu y
plt.legend(title='Jenis Kelamin', fontsize=12)  # Legenda
plt.show()  # Menampilkan plot
```

Bar Plot

```python
plt.figure(figsize=(10, 6))  # Mengatur ukuran gambar
sns.countplot(data=data, x='race/ethnicity', hue='test preparation course', palette='Set1')  # Membuat bar plot
plt.title('Bar Plot Kursus Persiapan Tes berdasarkan Etnis', fontsize=16)  # Judul plot
plt.xlabel('Etnis', fontsize=14)  # Label sumbu x
plt.ylabel('Jumlah', fontsize=14)  # Label sumbu y
plt.legend(title='Kursus Persiapan Tes', fontsize=12)  # Legenda
plt.show()  # Menampilkan plot
```

Heat Map

```python
correlation_matrix = data.corr(numeric_only=True)  # Menghitung matriks korelasi hanya pada kolom numerik
plt.figure(figsize=(10, 6))  # Mengatur ukuran gambar
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')  # Membuat heatmap
plt.title('Heatmap Matriks Korelasi', fontsize=16)  # Judul plot
plt.show()  # Menampilkan plot
```

Line Plot

```python
plt.figure(figsize=(10, 6))  # Mengatur ukuran gambar
data['math score'].rolling(window=20).mean().plot(color='purple')  # Membuat line plot rolling mean
plt.title('Grafik Garis Mean Bergulir Skor Matematika', fontsize=16)  # Judul plot
plt.xlabel('Indeks', fontsize=14)  # Label sumbu x
plt.ylabel('Skor Matematika (Rolling Mean)', fontsize=14)  # Label sumbu y
plt.show()  # Menampilkan plot
```

Distributions

```python
plt.figure(figsize=(10, 6))  # Mengatur ukuran gambar
sns.histplot(data=data, x='writing score', kde=True, bins=30, color='green')  # Membuat histogram plot
plt.title('Distribusi Skor Menulis', fontsize=16)  # Judul plot
plt.xlabel('Skor Menulis', fontsize=14)  # Label sumbu x
plt.ylabel('Frekuensi', fontsize=14)  # Label sumbu y
plt.show()  # Menampilkan plot
```
