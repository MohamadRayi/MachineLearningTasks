Langkah yang dapat diambil untuk melakukan Exploratory Data Analysis (EDA) dengan DataFrame yang dimiliki:

1.Periksa Data Awal:

Anda telah mendapatkan informasi ringkas tentang DataFrame dari output yang Anda berikan. Periksa kolom-kolom tersebut untuk memahami jenis data apa yang Anda miliki.

2.Periksa Tipe Data:

Periksa tipe data dari setiap kolom. Pastikan bahwa tipe data yang diparse sesuai dengan yang Anda harapkan. Misalnya, pastikan bahwa kolom "Age" yang seharusnya berisi angka memiliki tipe data numerik, bukan objek.

3.Periksa Data yang Hilang:

Periksa apakah ada nilai yang hilang (null) dalam DataFrame. Hal ini penting karena data yang hilang dapat memengaruhi analisis Anda.

4.Statistik Deskriptif:

Hitung statistik deskriptif untuk kolom-kolom numerik seperti mean, median, dan standar deviasi. Ini akan memberikan pemahaman awal tentang distribusi data.

5.Visualisasi Data:

Buat visualisasi data untuk masing-masing kolom, terutama kolom-kolom kategorikal. Anda dapat menggunakan histogram, box plot, bar plot, dan visualisasi lainnya untuk memahami distribusi dan pola data.

6.Analisis Korelasi:

Lakukan analisis korelasi antara kolom-kolom numerik untuk melihat hubungan antara mereka. Anda dapat menggunakan heatmap korelasi untuk memvisualisasikan ini.

7.Analisis Variabel Kategorikal:

Untuk kolom-kolom kategorikal, lakukan analisis untuk melihat frekuensi masing-masing kategori dan bagaimana mereka terkait dengan variabel lain.

8.Segmentasi Data:

Jika memungkinkan, lakukan segmentasi data berdasarkan kolom-kolom tertentu (misalnya, posisi pekerjaan, teknologi utama, atau ukuran perusahaan) dan lakukan analisis lebih lanjut untuk setiap segmen.

9.Identifikasi Outliers:

Periksa apakah ada outliers dalam data. Outliers dapat mempengaruhi hasil analisis Anda dan mungkin perlu ditangani secara khusus.

10.Analisis Time Series (Jika Ada):

Jika data memiliki dimensi waktu, lakukan analisis time series untuk memahami pola dan tren seiring waktu.

11.Uji Hipotesis (Jika Diperlukan):

Jika Anda memiliki pertanyaan tertentu atau hipotesis yang ingin Anda uji, lakukan uji hipotesis untuk mendapatkan pemahaman yang lebih dalam tentang data Anda.

**5.Visualisasi Data**

```python
# Visualisasi Histogram untuk kolom numerik
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()

# Daftar gaya yang berbeda untuk setiap plot
styles = ["whitegrid", "darkgrid", "ticks", "dark", "white"]

for i, col in enumerate(numeric_cols):
    # Set style untuk setiap iterasi
    sns.set_style(styles[i % len(styles)])

    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], bins=20, kde=True, color='skyblue')
    plt.title(col)
    plt.xlabel('Nilai')
    plt.ylabel('Frekuensi')
    plt.show()

#Membuat histogram untuk setiap kolom numerik dalam DataFrame.
#Histogram digunakan untuk memvisualisasikan distribusi data.
```

Untuk Bar plot ini Sengaja tidak ada output dikarenakan akan eror saat akan masuk ke github. run di dalam collab agar bisa masuk ke dalam github

```python
# Bar plot untuk kolom kategorikal
categorical_cols = df.select_dtypes(include='object').columns.tolist()
for col in categorical_cols:
    plt.figure(figsize=(12, len(df[col].unique()) * 0.5))  # Menyesuaikan tinggi gambar dengan jumlah kategori
    ax = sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.title(f'Distribusi Frekuensi {col}', fontsize=16, fontweight='bold')
    plt.xlabel('Frekuensi', fontsize=14)
    plt.ylabel(col, fontsize=14)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.grid(axis='x', linestyle='--', alpha=0.7)

    # Menambahkan label frekuensi pada bar plot
    total = len(df[col])
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_width() / total)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height() / 2
        ax.annotate(percentage, (x, y), fontsize=12, color='black', va='center')

    plt.tight_layout()
    plt.show()

#Membuat bar plot untuk setiap kolom kategorikal dalam DataFrame.
#Bar plot digunakan untuk memvisualisasikan frekuensi masing-masing kategori dalam kolom.
```

**7.Analisis Variabel Kategorikal**

```python
# List kolom kategorikal
categorical_cols = [
    'Gender',
    'City',
    'Position ',
    'Total years of experience',
    'Years of experience in Germany',
    'Seniority level',
    'Your main technology / programming language',
    'Other technologies/programming languages you use often',
    'Yearly bonus + stocks in EUR',
    'Annual bonus+stocks one year ago. Only answer if staying in same country',
    'Number of vacation days',
    'Employment status',
    'Сontract duration',
    'Main language at work',
    'Company size',
    'Company type',
    'Have you lost your job due to the coronavirus outbreak?',
    'Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'
]

# Plot frekuensi untuk setiap kolom kategorikal
plt.figure(figsize=(15, 60))
for i, col in enumerate(categorical_cols, 1):
    try:
        # Dapatkan frekuensi setiap nilai dan urutkan
        value_counts = df[col].value_counts().sort_values(ascending=False)

        # Gambarkan plot
        plt.subplot(len(categorical_cols), 1, i)
        sns.barplot(y=value_counts.index, x=value_counts.values)
        plt.title(f'Frekuensi {col}')
        plt.xlabel('Frekuensi')
        plt.ylabel(col)
    except ValueError as e:
        print(f"Error plotting {col}: {e}")

plt.tight_layout()
plt.show()
```

**8.Segmentasi Data**

```python
# Tentukan rentang gaji untuk segmentasi
salary_bins = [0, 30000, 50000, 80000, 120000, np.inf]
salary_labels = ['<30k', '30k-50k', '50k-80k', '80k-120k', '120k+']

# Tambahkan kolom baru untuk kategori gaji
df['Salary Category'] = pd.cut(df['Yearly brutto salary (without bonus and stocks) in EUR'], bins=salary_bins, labels=salary_labels)

# Hitung jumlah responden dalam setiap kategori gaji
salary_counts = df['Salary Category'].value_counts()

# Visualisasikan dengan diagram batang
plt.figure(figsize=(8, 6))
salary_counts.plot(kind='bar', color='skyblue')
plt.title('Jumlah Responden berdasarkan Kategori Gaji')
plt.xlabel('Kategori Gaji')
plt.ylabel('Jumlah Responden')
plt.xticks(rotation=45)
plt.show()
```

**9.Identifikasi Outliers**

```python
# Identifikasi outliers menggunakan metode IQR
numeric_df = df.select_dtypes(include=np.number)  # Pilih kolom-kolom numerik
Q1 = numeric_df.quantile(0.25)
Q3 = numeric_df.quantile(0.75)
IQR = Q3 - Q1

# Hitung batas atas dan bawah untuk identifikasi outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifikasi outliers dalam DataFrame numerik
outliers = ((numeric_df < lower_bound) | (numeric_df > upper_bound)).any(axis=1)

# Tampilkan jumlah outliers
print("\nJumlah Outliers:", outliers.sum())

# Visualisasikan outliers dengan scatterplot
plt.figure(figsize=(10, 6))

# Scatterplot untuk data outliers
sns.scatterplot(
    data=df[outliers],  # Pilih baris dengan setidaknya satu outlier
    color='red',  # Gunakan warna merah untuk menandai outliers
    s=100,  # Ukuran titik scatterplot
    marker='o'  # Tanda marker untuk outlier (lingkaran)
)

# Judul dan label sumbu
plt.title('Identifikasi Outliers menggunakan Metode IQR')
plt.xlabel('Index Data')
plt.ylabel('Value')

# Tampilkan scatterplot
plt.show()
```

**10.Analisis Time Series (Jika ada)**

```python
# Misalnya jika ada kolom 'Timestamp' sebagai dimensi waktu
if 'Timestamp' in df.columns:
    df['Timestamp'] = pd.to_datetime(df['Timestamp']) # Pastikan kolom 'Timestamp' sudah dalam format datetime
    df.set_index('Timestamp', inplace=True) # Set kolom 'Timestamp' sebagai index untuk analisis time series
    # Contoh analisis time series (misalnya tren gaji seiring waktu)
    plt.figure(figsize=(10, 6))
    df['Yearly brutto salary (without bonus and stocks) in EUR'].plot()
    plt.title('Tren Gaji seiring Waktu')
    plt.xlabel('Timestamp')
    plt.ylabel('Yearly brutto salary (without bonus and stocks) in EUR')
    plt.show()

#Melakukan analisis time series jika kolom 'Timestamp' tersedia dalam dataset.
#Memvisualisasikan tren gaji seiring waktu menggunakan plot time series.
```

11.Uji Hipotesis (Jika diperlukan)

```python
# Pisahkan data menjadi dua kelompok berdasarkan kolom "Years of experience in Germany"
experience_in_germany = df[df['Years of experience in Germany'].notnull()]
no_experience_in_germany = df[df['Years of experience in Germany'].isnull()]

# Lakukan uji t-test
t_stat, p_value = ttest_ind(experience_in_germany['Yearly brutto salary (without bonus and stocks) in EUR'],
                             no_experience_in_germany['Yearly brutto salary (without bonus and stocks) in EUR'],
                             equal_var=False)

# Tampilkan hasil uji hipotesis
print("Hasil Uji T-Test:")
print(f"Nilai T-Statistic: {t_stat}")
print(f"Nilai P-Value: {p_value}")

# Interpretasi hasil uji hipotesis
alpha = 0.05  # Tingkat signifikansi
if p_value < alpha:
    print("Karena nilai p-value kurang dari alpha (0.05), kita menolak hipotesis nol.")
    print("Artinya, terdapat perbedaan yang signifikan dalam rata-rata gaji tahunan antara responden yang memiliki pengalaman kerja di Jerman dan yang tidak memiliki pengalaman kerja di Jerman.")
else:
    print("Karena nilai p-value lebih besar dari alpha (0.05), kita tidak dapat menolak hipotesis nol.")
    print("Artinya, tidak terdapat cukup bukti statistik untuk menyimpulkan adanya perbedaan dalam rata-rata gaji tahunan antara kedua kelompok.")

# Kelompokkan data berdasarkan 'Years of experience in Germany' dan ambil rata-rata gaji tahunan
grouped_df = df.groupby('Years of experience in Germany')['Yearly brutto salary (without bonus and stocks) in EUR'].mean().reset_index()

# Visualisasikan perbedaan antara dua kelompok menggunakan box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Years of experience in Germany', y='Yearly brutto salary (without bonus and stocks) in EUR', data=grouped_df)
plt.title('Perbandingan Gaji Tahunan antara Kelompok dengan dan tanpa Pengalaman Kerja di Jerman')
plt.xlabel('Pengalaman Kerja di Jerman')
plt.ylabel('Gaji Tahunan (EUR)')
plt.show()
```
